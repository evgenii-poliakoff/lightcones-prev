# Tracing out the backward light cone

## Invariance of entanglement across the backward light cone boundary

Suppose we are interested in the dynamics only on the time interval $\left[0,T\right]$.

As discussed in the previous section {ref}`backward_lightcone`, we have the backward Lieb-Robinson light cone:

```{figure} outgoing_modes.png
:name: bck_lc
:width: 50%
```

The region colored with red consists of modes with are irreversibly decoupled from the local interacting site.

The total many-body state at time $t$ is $\left| \Psi \left(t\right) \right\rangle$.

We make the following bipartition: 

$$
    \left| \Psi \left(t\right) \right\rangle = \sum_p c_p \left| \psi_p \left(t\right) \right\rangle_{q+inside} \otimes 
    \left| \phi_p \left(t\right) \right\rangle_{outside}, 
$$ (in_out_bipart)


where "q+inside" refers to the modes of the open system (atom) and the bath modes inside the backward lightcone;
"outside" refers to the bath modes outside the backward lightcone (red colored area). 

We can select $\left| \phi_p \left(t\right) \right\rangle_{outside}$ as orthogonal basis elements, 

$$
    \left\langle \phi_q \right| \left. \phi_p \right \rangle_{outside} = \delta_{qp}.
$$

The corresponding wavefunctions for the inside part $\left| \psi_p \left(t\right) \right\rangle_{q+inside}$ need not orthogonal, but can be chosen to be normalized,

$$
     \left\langle \psi_p \right| \left. \psi_p \right \rangle_{q+inside} = 1.
$$

The many-body state state at a later time $\left| \Psi \left(t^{\prime}\right) \right\rangle$ is related by evolution operator,

$$
    \left| \Psi \left(t^{\prime}\right) \right\rangle = \widehat{U}\left(t^{\prime}, t \right)
          \left| \Psi \left(t\right) \right\rangle.
$$

Since the outgoing (outside the backward lightcone) modes have negligible average intensity of coupling, the evolution operator has the effective bipartitioning

$$
    \widehat{U}\left(t^{\prime} -t \right) = \widehat{U}_{q+inside}\left(t^{\prime}, t \right) \otimes \widehat{1}_{outside} + O\left(a_{cut}\right),
$$


that is it does not couple the outgoing mode.

Therefore, we have at a  later time:


$$
    \left| \Psi \left(t^\prime\right) \right\rangle = \sum_p c_p \left| \psi_p \left(t^\prime\right) \right\rangle_{q+inside} \otimes 
    \left| \phi_p \left(t\right) \right\rangle_{outside} + O\left(a_{cut}\right),. 
$$ (in_out_bipart_prime)

with

$$
    \left|\psi_p \left(t^\prime\right) \right\rangle_{q+inside} = \widehat{U}_{q+inside}\left(t^{\prime}, t \right) \left| \psi_p \left(t\right) \right\rangle_{q+inside}.
$$

Comparing the equations {eq}`in_out_bipart` and {eq}`in_out_bipart_prime`, we see that this bipartition is invariant under the future quantum motion.

The coefficients $c_p$ are time independent.

## Schmidt decomposition

We compute the density matrix of the quantum field outside the lightcone:

$$
    \widehat{\rho}_{outside} = \textrm{Tr}_{q+inside}\left[\left|\Psi\left(t\right)\right\rangle\left\langle\Psi\left(t\right)\right|\right]
$$

and find its eigenvalues:

$$
    \widehat{\rho}_{outside} \left| \phi_p \left(t\right) \right\rangle_{outside} = \left|c_p\right|^2 \left| \phi_p \left(t\right) \right\rangle_{outside}.
$$

The resulting $\left| \phi_p \left(t\right) \right\rangle_{outside}$ yield the bipartition of the form {eq}`in_out_bipart`, with the additional property that  $\left| \psi_p \left(t\right) \right\rangle_{q+inside}$ are also orthonormal. This is the Schmidt decomposition.

One can introduce a statistical cut off $\pi_{cut}$ so that we keep only a finite number of terms in {eq}`in_out_bipart` such that 

$$
    \left|c_p\right|^2  > \pi_{cut}.
$$

Suppose there are $m$ such terms.


## Computing the observables
 

Suppose we are interested in some observable $\widehat{O}_{q+inside} \otimes \widehat{O}_{outside}$, and want to compute its average $\overline{\widehat{O}_{q+inside} \otimes \widehat{O}_{outside}\left(t\right)}$ at later time moments $t^\prime\geq t$.


Then we form the $m\times m$ matrix 

$$
    O\left(p,q\right)_{outside} =  \left\langle \psi_p \right| \widehat{O}_{outside} \left| \psi_q \right \rangle_{outside}.
$$

We find its eigenvalues

$$
    \sum_{q=1}^m O\left(p,q\right)_{outside} f^{(l)}_q = O_l f^{(l)}_p.
$$

We choose a novel basis for bipartition:

$$
        \left| \psi^{(O)}_l \right \rangle_{outside} = \sum_{p=1}^m f^{(l)}_p \left| \psi_p \right \rangle_{outside}.
$$ (diagonalizing_basis)

 Then the average becomes

$$
    \overline{\widehat{O}_{q+inside} \otimes \widehat{O}_{outside}\left(t\right)} = \sum_p \left|c_p\right|^2  \left\langle \psi_p\left(t^\prime\right) \right| \widehat{O}_{q+inside}\left| \psi_p \left(t^\prime\right) \right \rangle_{q+inside} O_p
$$

for all later time moments $t^\prime\geq t$.

## Stochastic interpretation

We can compute this average at all times $t^\prime \geq t$ via the following Monte Carlo procedure:

- Given the state $\left| \Psi \left(t\right) \right\rangle$ at time $t$, we select the  basis  {eq}`diagonalizing_basis`

- We consider $\left|c_p\right|^2$ as probabilities $P\left(p\right)$ of events $p=1\ldots m$. Each event corresponds to a collapse

$$
    \left| \Psi \left(t\right) \right\rangle \to \left|\psi_p \left(t\right) \right\rangle_{q+inside}.
$$

Observe that after the collapse there is no modes outside the backward light cone. So the collapse is accompanied with reduction of the Hilbert space.

- We sample events $p$ according to their probability $P\left(p\right)$ via Monte Carlo algorithm. 

- As a result, we get samples of $\left|\psi_{p_s} \left(t\right) \right\rangle_{q+inside}$, $s=1\ldots M$, where $M$ is the number of samples. 
- Each sample is propagated to later times via

$$
    \left|\psi_{p_s} \left(t^\prime\right) \right\rangle_{q+inside} = \widehat{U}_{q+inside}\left(t^{\prime}, t \right)  \left|\psi_{p_s} \left(t\right) \right\rangle_{q+inside}
$$ (between_jumps)

- The sample average is computed as

$$
    A = \frac{1}{M} \sum_{s=1}^{M} \left\langle \psi_{p_s} \right| \widehat{O}_{q+inside} \left| \psi_{p_s} \right \rangle_{q+inside} O_{p_s}
$$


## Benefits of stochastic interpretation

- We avoid density matrices. The latter would be obtained if we were to trace out the modes outside the backward light cone. Instead, we stay on the level of wavefunctions.
- The proposed algorithm is massively parallel: each sample $\left|\psi_{p_s} \left(t^\prime\right) \right\rangle_{q+inside}$ is propagated independently. Different samples  can be computed in separate threads, processes, or even on separate servers. 

## Recurrent application


As we see from the figure {numref}`bck_lc`, the backward light cone shrinks as we move forward in time: new modes fall out of the light cone interior.

Then the above procedure can be applied recurrently: we get a sequence of quantum jumps:

$$
 \ldots \to \left|\psi_{p_1} \left(t_1\right) \right\rangle_{q+inside1} \to \left|\psi_{p_2} \left(t_2\right) \right\rangle_{q+inside2} \to \left|\psi_{p_3} \left(t_3\right) \right\rangle_{q+inside3} \to \ldots  
$$

at a sequence of times $\ldots, t_1, t_2, t_3 \ldots$. Between the quantum jumps the samples are propagated unitarily according to {ref}`between_jumps`.
